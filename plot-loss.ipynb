# Plot losses

```python
import re
import matplotlib.pyplot as plt
```

```python
pattern = re.compile(r'\(epoch: (\d+), iters: (\d+), time: (\d+\.\d+), data: (\d+\.\d+)\) D: (\d+\.\d+) G_A: (\d+\.\d+) cycle_A: (\d+\.\d+) idt_A: (\d+\.\d+) G_B: (\d+\.\d+) cycle_B: (\d+\.\d+) idt_B: (\d+\.\d+)')
loss_file = "/home/tuur/Downloads/loss_log.txt"
```

```python
with open(loss_file, 'r') as f:
    data = f.read().strip().splitlines()
```

```python
epochs = []
D = []
G_A = []
G_B = []
cycle_A = []
cycle_B = []
idt_A = []
idt_B = []

for line in data:
    match = re.match(pattern, line)
    if match: 
        print(line)
        epochs.append(int(match.group(1)))
        D.append(float(match.group(5)))
        G_A.append(float(match.group(6)))
        cycle_A.append(float(match.group(7)))
        idt_A.append(float(match.group(8)))
        G_B.append(float(match.group(9)))
        cycle_B.append(float(match.group(10)))
        idt_B.append(float(match.group(11)))
```

```python
plt.plot(epochs, idt_A)
plt.plot(epochs, idt_B)
plt.show()
```

```python
plt.plot(epochs, G_A)
plt.plot(epochs, G_B)
plt.show()
```

```python
plt.plot(epochs, D)
plt.plot()
```
